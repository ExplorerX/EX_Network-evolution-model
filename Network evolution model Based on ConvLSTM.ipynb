```python
data1 = pd.read_csv("source_data.csv", header=None, sep=',')
data2 = pd.read_csv("source_data.csv", header=None, sep=',')
```

```python
data1.iloc[0]
```

```python
complete_data = []
n_source_data = data1.shape[0]
for index, row in data1.iterrows():
    if row[3] == "up":
        start = row[0]
        node1 = row[1]
        node2 = row[2]
        for i in range(index+1, n_source_data):
            if (data1.iloc[i][3] == "down") and (node1 == data1.iloc[i][1]) and (node2 == data1.iloc[i][2]):
                end = data1.iloc[i][0]
                continuance = end - start
                print([node1, node2, start, end, continuance])
                complete_data.append([node1, node2, start, end, continuance])
                break
```
